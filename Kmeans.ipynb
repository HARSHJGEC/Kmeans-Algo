```python
import pandas as pd
import numpy as np
import random as rd
import math
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=pd.read_csv('cancer.csv')
df.drop(columns=["Unnamed: 32","id","diagnosis"],inplace=True)

```

```python
for i in range(0,df.shape[1]):
    col=df.iloc[:,i]
    for j in range(0,df.shape[0]):
        col.iloc[j]=(col.iloc[j]-col.min())/(col.max()-col.min())

data=df.values
k=2
```

```python
cluster=rd.sample(range(data.shape[0]),k)
cluster
```

```python
def equi_dist(cluster,arr)->int:
    dist=0
    for i in range(0,cluster.shape[0]):
        temp=cluster[i]-arr[i]
        temp=temp**2
        dist=dist+temp
        result=math.sqrt(dist)
    return result
```

```python
def centroid(cluster_list):
    points=[]
    for j in range(0,k):
        df_temp=df.copy()
        set1=set(cluster_list[j])
        for i in range(0,df.shape[0]):
            if i  not in set1: 
                df_temp=df_temp.drop([i])
        temp=[]
        for i in range(0,df.shape[1]):
            col=df_temp.iloc[:,i]
            temp.append(col.sum()/col.shape[0])
        points.append(temp)
        del set1
    cluster1=np.array(points)
    return cluster1
```
